```python
!pip install pyngrok
# 需要安裝的套件
from diffusers import StableDiffusionPipeline
from flask import Flask, request, jsonify
import base64
from io import BytesIO
from PIL import Image
#import os
from pyngrok import ngrok
import atexit

app = Flask(__name__)

# 初始化 Stable Diffusion
model_id = "" #請換成你自己的模型名稱
pipe = StableDiffusionPipeline.from_pretrained(
    model_id
)
#diffusers reference: https://huggingface.co/docs/diffusers/main/en/quicktour
#from_pretrained reference: https://huggingface.co/docs/diffusers/v0.32.2/en/api/models/overview#diffusers.ModelMixin.from_pretrained
#pipeline reference: https://huggingface.co/docs/diffusers/main/en/api/pipelines/overview

pipe = pipe.to("cuda")

@app.route('/',methods=['POST'])
def generate_image():
    data = request.json
    prompt = data.get('prompt', '')

    # 生成圖片
    image = pipe(prompt).images[0]

    # 將圖片轉換為 base64
    buffered = BytesIO()
    image.save(buffered, format="PNG")
    img_str = base64.b64encode(buffered.getvalue()).decode()

    return jsonify({
        'status': 'success',
        'image': img_str
    })

def start_server():
    # 設定 ngrok token

    # 啟動 ngrok tunnel

    #ngrok reference:https://ngrok.github.io/ngrok-python/

    # 啟動 Flask
    app.run(port=5000)

if __name__ == '__main__':
    start_server()
```
